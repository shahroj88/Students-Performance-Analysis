# 2.	Central-Tendency Metrics

**	Compute mean, median, and mode for each of the three score columns.**



```
df = pd.read_csv("StudentsPerformance.csv")

scores = df[['math score', 'reading score', 'writing score']]

mean_scores = scores.mean()

median_scores = scores.median()

mode_scores = scores.mode()

print("Mean Scores:\n", mean_scores)
print("\nMedian Scores:\n", median_scores)
print("\nMode Scores:\n", mode_scores)

```

# 3.	Dispersion & Shape

**	Calculate standard deviation, variance, and coefficient of variation for math score.**

**	Determine skewness and kurtosis for reading score (use built-in functions or manual formulas).**


```
df = pd.read_csv("StudentsPerformance.csv")

math_std = df['math score'].std()
math_var = df['math score'].var()
math_cv = (math_std / df['math score'].mean()) * 100

print("Math Score Statistics:")
print("Standard Deviation:", math_std)
print("Variance:", math_var)
print("Coefficient of Variation (%):", math_cv)

reading_skew = df['reading score'].skew()
reading_kurt = df['reading score'].kurtosis()

print("\nReading Score Distribution Shape:")
print("Skewness:", reading_skew)
print("Kurtosis:", reading_kurt)
```

# 4.	Correlation & Covariance

**-	Produce a correlation matrix for the three scores and identify the strongest pairwise relationship. **

**-	Explain in one sentence how covariance differs from correlation and why we prefer the latter when comparing variables with different units. **


```

df = pd.read_csv("StudentsPerformance.csv")

scores = df[['math score', 'reading score', 'writing score']]

corr_matrix = scores.corr()
print("Correlation Matrix:\n")
print(corr_matrix)

strongest = corr_matrix.unstack().sort_values(ascending=False)
strongest = strongest[strongest < 1].head(1)

print("\nStrongest Pairwise Relationship:\n")
print(strongest)
```

# 5.	Basic Probability

**What is the probability that a randomly selected student scored ≥ 80 in both math and reading?**
** P(A∩B)=count of students with both scores ≥ 80total studentsP(A \cap B) = \dfrac{\text{count of students with both scores ≥ 80}}{\text{total students}}P(A∩B)=total studentscount of students with both scores ≥ 80 **


```

df = pd.read_csv("StudentsPerformance.csv")

count_both = df[(df['math score'] >= 80) & (df['reading score'] >= 80)].shape[0]

total_students = df.shape[0]
probability = count_both / total_students

print("Students with Math ≥80 and Reading ≥80:", count_both)
print("Total Students:", total_students)
print("Probability P(A ∩ B):", probability)
```

# 6.	Visual Checks

**	Plot a histogram (or box-plot) of math scores and briefly comment on the distribution shape.**



```

df = pd.read_csv("StudentsPerformance.csv")

plt.hist(df['math score'], bins=10, edgecolor='black', color='lightcoral')

plt.xlabel("Math Score")
plt.ylabel("Frequency")
plt.title("Distribution of Math Scores")

plt.show()
```

```
df = pd.read_csv("StudentsPerformance.csv")

plt.boxplot(
    df['math score'],
    patch_artist=True,
    boxprops=dict(facecolor='skyblue', color='blue'),
    medianprops=dict(color='red', linewidth=2),
    whiskerprops=dict(color='black'),
    capprops=dict(color='black'),
    flierprops=dict(marker='o', markerfacecolor='orange', markersize=6)
)

plt.ylabel("Math Score")
plt.title("Box Plot of Math Scores")

plt.show()

```

```
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("StudentsPerformance.csv")

s = df['math score'].sort_values().reset_index(drop=True)
s = s.rolling(100).mean()

plt.plot(s, color='green')
plt.xlabel("Students (sorted)")
plt.ylabel("Math Score")
plt.title("Smooth Trend of Math Scores")
plt.show()
```

**The average (mean and median) scores of math, reading, and writing are close to each other, which shows that students perform almost equally in all subjects. The spread of math scores is moderate, meaning students’ scores are not too far apart. Most students scored around the middle range. The relationship between math and reading scores is positive, which means students who score well in math usually also score well in reading. Overall, the data shows a balanced and consistent performance among students.**
